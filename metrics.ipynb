```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('step_metrics.csv')

# Plot each metric
plt.figure(figsize=(12, 8))

# Plot train loss
plt.subplot(2, 2, 1)
plt.plot(df['step'], df['train_loss'], color='blue')
plt.title('Train Loss')
plt.xlabel('Step')
plt.ylabel('Loss')

# Plot train accuracy
plt.subplot(2, 2, 2)
plt.plot(df['step'], df['train_accuracy'], color='green')
plt.title('Train Accuracy')
plt.xlabel('Step')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Histograms or Density Plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['train_loss'], kde=True, color='blue')
plt.title('Distribution of Train Loss')

plt.subplot(1, 2, 2)
sns.histplot(df['train_accuracy'], kde=True, color='green')
plt.title('Distribution of Train Accuracy')

plt.tight_layout()
plt.show()

# Box Plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y='train_loss', data=df, color='blue')
plt.title('Box Plot of Train Loss')

plt.subplot(1, 2, 2)
sns.boxplot(y='train_accuracy', data=df, color='green')
plt.title('Box Plot of Train Accuracy')

plt.tight_layout()
plt.show()
```
